```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

from plottable import ColumnDefinition, Table
from plottable.cmap import normed_cmap
from plottable.plots import image
```

```python
df = pd.read_html('https://fbref.com/en/comps/189/Womens-Super-League-Stats', attrs={"id": "results2023-20241891_overall"})[0]
```

```python
df.head()
```

```python
df.columns
```

```python
df = df[[
    'Rk', 'Squad', 'MP', 'W', 'D', 'L', 'GF', 'GA', 'GD', 'Pts'
]]
```

```python
df.columns
```

```python
bg_color = "#FFFFFF"
text_color = "#000000"

row_colors = {
    "top4": "#E1FABC",
    "top6": "#FFFC97",
    "relegation": "#E79A9A",
    "even": "E2E2E1",
    "odd": "#B3B0B0",
}

plt.rcParams["text.color"] = text_color
plt.rcParams["font.family"] = "monospace"
```

```python
col_defs = [
    ColumnDefinition(
            name="Rk",
            textprops={"ha": "center"},
            width=0.5,
        ),
     ColumnDefinition(
            name="Squad",
            textprops={"ha": "left", "va": "center", "weight": "bold"},
            width=1.75,
        ),
     ColumnDefinition(
            name="MP",
            group="Matches Played",
            textprops={"ha": "center"},
            width=0.5,
        ),
    ColumnDefinition(
            name="W",
            group="Won",
            textprops={"ha": "center"},
            width=0.5,
        ),
    ColumnDefinition(
            name="D",
            group="Draw",
            textprops={"ha": "center"},
            width=0.5,
        ),
    ColumnDefinition(
            name="L",
            group="L",
            textprops={"ha": "center"},
            width=0.5,
        ),
    ColumnDefinition(
            name="GF",
            group="GF",
            textprops={"ha": "center"},
            width=0.5,
        ),
    ColumnDefinition(
            name="GA",
            group="GA",
            textprops={"ha": "center"},
            width=0.5,
        ),
    ColumnDefinition(
            name="GD",
            group="GD",
            textprops={"ha": "center"},
            width=0.5,
        ),
    ColumnDefinition(
            name="MP",
            group="MP",
            textprops={"ha": "center"},
            width=0.5,
        ),
    ColumnDefinition(
            name="Pts",
            group="Points",
            textprops={"ha": "center"},
            width=0.5,
        ),
]
```

```python
fig, ax = plt.subplots(figsize=(20,22))
fig.set_facecolor(bg_color)
ax.set_facecolor(bg_color)

table = Table(
    df,
    column_definitions = col_defs,
    index_col = "Rk",
    row_dividers = True,
    row_divider_kw={"linewidth": 1, "linestyle": (0, (1,5))},
    footer_divider = True,
    textprops = {"fontsize": 14},
    ax = ax
)

for idx in [0, 1, 2]:
    table.rows[idx].set_facecolor('#6DAEDB')
    
for idx in [11]:    
    table.rows[idx].set_facecolor('#C95D63')
    
```
