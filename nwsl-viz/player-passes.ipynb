```python
# create columns for player passes with passEndX and passEndY
df['end_x'] = None
df['end_y'] = None
df['receiver_player_name'] = None

# Iterate through each row
for index, row in df.iterrows():
    # If current action is a pass
    if row['typeId'] == 1:
        # Find the index of the next row with action
        next_index = df.index[index + 1]
        if next_index < len(df):
            # Update end locations
            df.at[index, 'end_x'] = df.at[next_index, 'x']
            df.at[index, 'end_y'] = df.at[next_index, 'y']
            if row['outcome'] == 1:
                # Add receiver player name
                df.at[index, 'receive_playerName'] = row['playerName']
# create a column to check for progressive pass
df['beginning'] = np.sqrt(np.square(105 - df['x']) + np.square(52.5 - df['y']))
df['end'] = np.sqrt(np.square(105 - df['end_x']) + np.square(52.5 - df['end_y']))

df['progressive'] = [(df['end'][x]) / (df['beginning'][x]) < 0.75 for x in range(len(df.beginning))]

df[df.typeId == 1]
```
