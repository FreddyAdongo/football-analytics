```python
import pandas as pd

from tabulate import tabulate
```

```python
df = pd.read_html('https://fbref.com/en/comps/189/Womens-Super-League-Stats',
                 attrs={'id': 'results2023-20241891_overall'})
```

```python
df = df[0]
```

```python
df.head()
```

```python
#sort by goal difference
df = df.sort_values('GD', ascending=False)
df.head(10)
```

```python
df = df[['Squad', 'GF', 'GA', 'GD', 'Pts']]

table = tabulate(df.head(10), headers='keys', tablefmt='pipe')
```

```python
table = tabulate(df.head(10), headers='keys', tablefmt='pipe')
print(table)
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()

ax.table(
    cellText=df.head(10).values, 
    colLabels=df.columns, 
    loc='center',
    cellLoc='center',
    edges='open',
    colWidths=[0.2, 0.1, 0.1, 0.1, 0.1],
    colColours=['blue']*5
)

ax.axis('off')
```
