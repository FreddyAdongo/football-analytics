```python
import pandas as pd
```

```python
# open match file
df = pd.read_csv('../match-data/nwsl-events/2025-04-13_Gotham_FC__vs__North_Carolina_Courage.csv')
```

```python
df.head()
```

```python
# get shots - there is 'Goal' - 16, 'Attempt Saved' - 15, 'Miss' - 13, 'Post' - 14
df[df.typeId == 16] # Goal
df[df.typeId == 15] # Attempt Saved
df[df.typeId == 13] # Miss
df[df.typeId == 14] # Post

# place shots in one dataframe
shot_type_ids = [13, 14, 15, 16] # Create a list of shot typeIds

shots_df = df[df['typeId'].isin(shot_type_ids)] # Filter the DataFrame in one operation

label = {13: 'Miss',
    14: 'Post',
    15: 'Saved',
    16: 'Goal'
        }
shots_df['shot_outcome'] = shots_df['typeId'].map(label)

shots_df
```
