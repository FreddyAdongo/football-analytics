```python
import pandas as pd
import numpy as np
```

```python
# open file and store in dataframe
df = pd.read_csv('../match-data/nwsl-events/2025-04-22_Portland_Thorns__vs__Gotham_FC.csv')
df.head()
```

```python
# get defensive actions, we need - tackles, clearances, interceptions, aerial, recoveries
df[df.typeId == 7] # tackle
df[df.typeId == 12] # clearance
df[df.typeId == 8] # interception
df[df.typeId == 44] # aerial
df[df.typeId == 49] # recovery

# drop unneccessary columns
df.drop(columns=['timeStamp', 'contestantId', 'lastModified', 'assist', 'keyPass', 'playerId'])

defensive_ids = [7, 12, 8, 44, 49]
defensive_df = df[df['typeId'].isin(defensive_ids)]
```
