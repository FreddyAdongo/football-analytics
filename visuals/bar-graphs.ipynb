```python
import pandas as pd
```

```python
url = 'https://fbref.com/en/squads/85c458aa/North-Carolina-Courage-Stats'

df = pd.read_html(url, attrs={'id' : 'stats_standard_182'})[0]
```

```python
df.head(10)
```

```python
df.tail(10)
```

```python
df = df[df.columns.drop(list(df.filter(regex='Per 90')))]
```

```python
df = df.droplevel(0, axis=1)
```

```python
df.tail(10)
```

```python
df.head(10)
```

```python
df.drop([22,23], axis=0, inplace=True)
df.tail()
```

```python
df = df.fillna(0)
df['Min'] = df['Min'].astype(int)
df['Gls'] = df['Gls'].astype(int)
df['Ast'] = df['Ast'].astype(int)
```

```python
df.head(10)
```

```python
df = df[df['Min'].astype(int) > 200]
```

```python
# Let's look at the columns we are interested in
df[['90s', 'Gls', 'Ast', 'G+A', 'xG', 'xAG', 'PrgC', 'PrgP', 'PrgR']].head(10)

# let's also make sure that the columns are of the correct type
df[['90s', 'xG', 'xAG']] = df[['90s', 'xG', 'xAG']].astype(float)
df[['Gls', 'Ast', 'G+A', 'PrgC', 'PrgP', 'PrgR']] = df[['Gls', 'Ast', 'G+A', 'PrgC', 'PrgP', 'PrgR']].astype(int)
```

```python
#calculate per 90s
df['goals_per_90'] = df['Gls'] / df['90s']
df['assists_per_90'] = df['Ast'] / df['90s']
df['goals_assists_per_90'] = df['G+A'] / df['90s']
df['expected_goals_per_90'] = df['xG'] / df['90s']
df['expected_assists_per_90'] = df['xAG'] / df['90s']
df['progressive_carries_per_90'] = df['PrgC'] / df['90s']
df['progressive_passes_per_90'] = df['PrgP'] / df['90s']
df['progressive_receptions_per_90'] = df['PrgR'] / df['90s']
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots()

ax.barh(df['Player'], df['expected_goals_per_90'], color='#104275')
ax.set_facecolor('#EFE9E6')
ax.set_xlabel("Expected Goals (xG)")
ax.set_title("NC Courage Expected Goals Generated | 2024 NWSL Season \nPlayers with over 200 minutes", va='center', y=1.05)

# Add spines
ax.spines["top"].set(visible = False)
ax.spines["right"].set(visible = False)

# Add grid and axis labels
ax.grid(True, color = "lightgrey", ls = ":")
```
