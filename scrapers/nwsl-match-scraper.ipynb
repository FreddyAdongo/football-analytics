```python
# clean up the match info and event data
def process_match_info(match_info, event_data):
    event_df = pd.json_normalize(event_data)

    home_team = None
    away_team = None
    home_contestant_id = None
    away_contestant_id = None

    # Get team names and contestant IDs
    for contestant in match_info.get('contestant', []):
        if contestant.get('position') == 'home':
            home_team = contestant.get('name')
            home_contestant_id = contestant.get('id')
        elif contestant.get('position') == 'away':
            away_team = contestant.get('name')
            away_contestant_id = contestant.get('id')
    
    # Add teams to event data
    if home_team and away_team:
        event_df['home_team'] = home_team
        event_df['away_team'] = away_team

    # Create Squad and Position columns if we have contestant IDs
    if home_contestant_id and away_contestant_id:
        # Create a dictionary mapping player IDs to their teams
        player_teams = {}
        
        # Map players to their teams
        for _, row in event_df.iterrows():
            if pd.notna(row.get('playerId')):
                if row['contestantId'] == home_contestant_id:
                    player_teams[row['playerId']] = home_team
                elif row['contestantId'] == away_contestant_id:
                    player_teams[row['playerId']] = away_team
        
        # Add Squad column
        event_df['Squad'] = event_df['playerId'].map(player_teams)
        
        # Add Position column (Home or Away)
        event_df['Position'] = event_df['Squad'].apply(
            lambda x: 'Home' if x == home_team else 'Away' if x == away_team else None
        )
        
        # For rows without playerId but with contestantId, assign team
        event_df.loc[
            (event_df['Squad'].isna()) & 
            (event_df['contestantId'] == home_contestant_id), 
            ['Squad', 'Position']
        ] = [home_team, 'Home']
        
        event_df.loc[
            (event_df['Squad'].isna()) & 
            (event_df['contestantId'] == away_contestant_id), 
            ['Squad', 'Position']
        ] = [away_team, 'Away']

    event_df.drop(columns=['home_team', 'away_team'], axis=1, inplace=True, errors='ignore')

    description = match_info.get('description', 'match')
    local_date = match_info.get('localDate', 'unknown_date')

    clean_description = (
        description.replace(' ', '_')
                  .replace('vs.', 'vs')
                  .replace('vs', '_vs_')
                  .replace('/', '_')
                  .replace(':', '')
    )

    filename = f'../match-data/nwsl-events/{local_date}_{clean_description}.csv'
    
    # Save to CSV
    event_df.to_csv(filename, index=False)
    print(f"Saved match data to: {filename}")
    
    return event_df
```
